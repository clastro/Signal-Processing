### Filtering for P-Wave

```python
from scipy import signal
```

```python
b, a = signal.butter(N = 2, Wn = [1,12], btype='band',fs=256) 
```

```python
medifilt = scipy.signal.medfilt(abs(sample[400:1100]),21)
```

```python
plt.plot(medifilt)
plt.show()
```

```python
filtered = signal.filtfilt(b,a,medifilt)
```

```python
np.argmax(filtered)
```

### Phasor Transform

```python
plt.plot(np.arctan(filtered))
plt.hlines(1.5,0,700,color="red")
plt.hlines(1,0,700,color="blue")
plt.show()
```
