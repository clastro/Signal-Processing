```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import electrocardiogram
```

```python
x = electrocardiogram()[1000:1600]
x
plt.plot(x)
```

```python
from scipy import signal
```

### Band Pass Filter

```python
b, a = signal.butter(N = 2, Wn = [0.5,150], btype='band',fs=305) 
#N : Order
#Wn은 fs의 1/2 내로
```

```python
filtered = signal.filtfilt(b,a,x)
```

```python
plt.plot(filtered)
```

### High Pass Filter

```python
b, a = signal.butter(N = 2, Wn = 5, btype='high',fs=200)
```

```python
filtered = signal.filtfilt(b,a,x)
plt.plot(filtered)
```

### Low Pass Filter

```python
b, a = signal.butter(N = 2, Wn = 10, btype='low',fs=100)
```

```python
filtered = signal.filtfilt(b,a,x)
plt.plot(filtered)
```
